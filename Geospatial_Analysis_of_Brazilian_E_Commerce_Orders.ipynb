##**Geospatial Analysis of Brazilian E-Commerce Orders**



```
import folium
import pandas as pd
```

```
folium.__version__
```

### Data set of orders made between 2016 and 2018 at Brazilian Market

```
df_orders = pd.read_csv("olist_geolocation_dataset.csv")
```

```
df_orders.head(10)
```

```
df_orders.shape
```

### Split dataset by states

```
UniqueStates = df_orders.geolocation_state.unique()
DataFrameDict = {elem : pd.DataFrame for elem in UniqueStates}
```

```
for state,data in df_orders.groupby('geolocation_state'):
  DataFrameDict[state] = data
```

```
len(DataFrameDict)
```

```
DataFrameDict['SP']
```

### Merging a fraction of each dataset into one sample dataset 

```
perc = 0.001
df_orders_samp = pd.DataFrame()
for x in DataFrameDict:
  lent = int(len(DataFrameDict[x])*perc)
  print(lent)
  DataFrameDict[x] = DataFrameDict[x].iloc[0:lent,:]
```

```
df_orders_samp = pd.concat(DataFrameDict)
```

```
df_orders_samp.shape
```

### Visualizing the sample dataset

```
lat = -14.2350 
lng = -51.9253
bz_map = folium.Map(location=[lat,lng], zoom_start=7, tiles='Stamen Terrain')
```

```
from folium import plugins

braz_map = folium.Map(location = [lat,lng], zoom_start = 4)

orders = plugins.MarkerCluster()

for lat,lng,zcode in zip(df_orders_samp.geolocation_lat,df_orders_samp.geolocation_lng,df_orders_samp.geolocation_zip_code_prefix):
  orders.add_child(
      folium.Marker(
          [lat,lng],
          popup=zcode
      )
  )
braz_map.add_child(orders)
```
