## **Collaborative-Based Recommendation System**

```
from scipy.stats import pearsonr
```

```
products_df.columns
```

```
collb_df = products_df.copy()
#drop categories and any other columns that are not needed
collb_df = collb_df.drop(columns=['categories','reviews.text'])
collb_df
```

```
userInput = [
             {'asins':'B01AHB9CN2','reviews.rating':2.0},
             {'asins':'B01AHB9CYG','reviews.rating':4.0},
             {'asins':'B018Y225IA','reviews.rating':3.0},
             {'asins':'B018Y229OU','reviews.rating':4.0}           
]
inputProducts = pd.DataFrame(userInput)
inputProducts
```

```
userSubset = collb_df[collb_df['asins'].isin(inputProducts['asins']).tolist()]
userSubset.sort_values('reviews.username').head(20)
```

```
userSubsetGroup = userSubset.groupby(['reviews.username'])
```

```
userSubsetGroup = sorted(userSubsetGroup,  key=lambda x: len(x[1]), reverse=True)
```

```
userSubsetGroup = userSubsetGroup[0:100]
```

## Apply Pearson Correlation and get weighted score

```
pearsonCorrelationDict = {}

for name, group in userSubsetGroup:
  group = group.sort_values(by='asins')
  temp_df = inputProducts[inputProducts['asins'].isin(group['asins'].tolist())]
  tempRatingList = temp_df['reviews.rating'].tolist()
  tempGroupList = group['reviews.rating'].tolist()
  pearsonCorrelationDict[name] = round(pearsonr(tempRatingList, tempGroupList)[0], 3)
```

```
pearsonCorrelationDict
```

```
pearsonDF = pd.DataFrame.from_dict(pearsonCorrelationDict, orient='index').reset_index(drop=False)
pearsonDF.columns = ['reviews.username', 'similarityIndex']
pearsonDF = pearsonDF.sort_values(by='similarityIndex', ascending=False)
pearsonDF
```

```
topUsers = pearsonDF.iloc[0:20]
```

```
topUsersRating = pd.merge(topUsers, products_df, left_on='reviews.username', right_on='reviews.username', how='inner')
topUsersRating
```

```
weightedRating = topUsersRating['similarityIndex']*topUsersRating['reviews.rating']
topUsersRating['weightedRating'] = weightedRating
topUsersRating.columns
```

```
tempTopUsersRating = topUsersRating.groupby('asins').sum()['similarityIndex']
topUsersRatingGrouped = topUsersRating.groupby('asins')
weightedRating = []
for sim, data in topUsersRatingGrouped:
  weightedRating.append(topUsersRatingGrouped.get_group(sim).sum()['weightedRating'])
```

```
tempTopUsersRating = topUsersRating.groupby('asins').sum()[['similarityIndex']]
tempTopUsersRating.columns = ['sum_similarityIndex']
tempTopUsersRating['sum_weightedRating'] = weightedRating
tempTopUsersRating.head()
```

## Recommendations

```
#Creates an empty dataframe
recommendation_df = pd.DataFrame()
#Now we take the weighted average
recommendation_df['recommendation score'] = tempTopUsersRating['sum_weightedRating']/tempTopUsersRating['sum_similarityIndex']
#recommendation_df['asins'] = tempTopUsersRating.index
recommendation_df.reset_index(inplace=True)
```

```
recommendation_df = recommendation_df.sort_values(by='recommendation score', ascending=False)
recommendation_df.head(20)
```

```
user_recommendationTable_df = pd.merge(products_df, recommendation_df).sort_values(by='recommendation score', ascending=False)
user_recommendationTable_df
```
